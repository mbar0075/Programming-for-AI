```python
#Imports
import numpy as np
import numpy.linalg as la
from sympy import *
```

```python
# 3.e

L = np.matrix(
    [
        [  0, 1/3,   0, 1/3, 1/3,   0], # A
        [  0,   0,   0,   0,   1,   0], # B
        [1/3, 1/3,   0, 1/3,   0,   0], # C
        [  0, 1/2, 1/2,   0,   0,   0], # D
        [  0,   0,   1,   0,   0,   0], # E
        [  0,   0,   0,   0,   0,   1]  # F
    ]
).T

d = 0.5
n = 6
J = np.ones([6, 6])

M = d * L + ((1 - d) / n) * J

# 3.f

r = (np.ones([1, 6]) / 6).T

prev_norm = la.norm(M@r - r)

while prev_norm > 0.005:
    prev_norm = la.norm(M@r - r)
    r = M@r

Matrix(r)
```

```python
# 4.g

# 4.g

from random import randint

def gen_net(n_pages):
    link_mat = np.zeros([n_pages, n_pages])

    for i in range(n_pages):
        div = randint(1, n_pages - 1)

        for _ in range(div):
            index = randint(0, n_pages - 1)
            
            while (link_mat[i][index] != 0) or (index == i):
                index = randint(0, n_pages - 1)

            link_mat[i][index] = 1 / div

    return link_mat.T
```

```python
# 4.h

def page_rank(link_mat, d):
    L = link_mat
    size = len(L)

    n = size
    J = np.ones([size, size])

    M = d * L + ((1 - d) / n) * J

    r = (np.ones([1, size]) / size).T

    prev_norm = la.norm(M@r - r)

    while prev_norm > 0.005:
        prev_norm = la.norm(M@r - r)
        r = M@r

    return r

Matrix(page_rank(gen_net(5), 0.5)) # Input a link matrix generated by the function gen_net(), damping of 0.5
```
